```python
# import packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the Yellow cabs data- January month

yellow_cabs_data_jan = pd.read_csv('Documents/yellow_tripdata_2018-01.csv')
```

```python
#read the taxi lookup data

taxi_data = pd.read_csv("Documents/taxi _zone_lookup.csv")
taxi_data
```

```python
# merging the data to get location id and borough
merged_data=pd.merge(yellow_cabs_data_jan, taxi_data, on = None, left_on = 'PULocationID', right_on = 'LocationID')
```

```python
#merged data - Jan

df_Jan = pd.DataFrame(merged_data)
df_Jan
```

```python
# To plot the result of Manhattan 
df_Jan = pd.DataFrame(merged_data)
data = df_Jan.loc[df_Jan['Borough']=='Manhattan', 'tpep_pickup_datetime':'tpep_dropoff_datetime']
df_Jan = pd.DataFrame(data)
result = []
for i in range(0,len(df_Jan)):
    t1 = pd.to_datetime(df_Jan.iloc[i,0])
    t2 = pd.to_datetime(df_Jan.iloc[i,1])
    result.append((pd.Timedelta(t2 - t1).seconds / 60.0))
result1 = pd.DataFrame(result)
values = result1.get_values()
values = pd.DataFrame(values)
Manhattan = pd.DataFrame(values)
Manhattan.plot()
```
