```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the yellowcab data of january month 
yellow_cabs_data = pd.read_csv('Documents/yellow_tripdata_2018-01.csv')
```

```python
# reading the taxi zone lookup data
taxi_data = pd.read_csv("Documents/taxi _zone_lookup.csv")
taxi_data
```

```python
# merging the data to get the locationid and borough details

merged_data=pd.merge(yellow_cabs_data, taxi_data, on = None, left_on = 'PULocationID', right_on = 'LocationID')
```

```python
df_Jan = pd.DataFrame(merged_data)
```

```python
df_Jan
```

```python
# To get the average number of trips recoreded over the Borough - Manhattan for january month
df_Man_Jan = df_Jan.loc[(df_Jan['Borough']=='Manhattan')  & (df_Jan['tpep_pickup_datetime'].str.startswith('2018-01'))]
av_Man_Jan = len(df_Man_Jan)/31
av_Man_Jan
```

```python
#To get the average number of trips recoreded over the Borough - EWR for january month
df_EWR_Jan = df_Jan.loc[(df_Jan['Borough']=='EWR')  & (df_Jan['tpep_pickup_datetime'].str.startswith('2018-01'))]
av_EWR_Jan = len(df_EWR_Jan)/31
av_EWR_Jan
```

```python
#To get the average number of trips recoreded over the Borough - Queens for january month
df_Queens_Jan = df_Jan.loc[(df_Jan['Borough']=='Queens')  & (df_Jan['tpep_pickup_datetime'].str.startswith('2018-01'))]
av_Queens_Jan = len(df_Queens_Jan)/31
av_Queens_Jan
```

```python
#To get the average number of trips recoreded over the Borough - Staten island for january month
df_Staten_Jan = df_Jan.loc[(df_Jan['Borough']=='Staten Island')  & (df_Jan['tpep_pickup_datetime'].str.startswith('2018-01'))]
av_Staten_Jan = len(df_Staten_Jan)/31
av_Staten_Jan
```
