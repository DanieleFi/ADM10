```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from pylab import rcParams
import seaborn as sb

import scipy 
from scipy.stats import spearmanr
```

```python
#yellow cabs data - january
yellow_cabs_data = pd.read_csv('Documents/yellow_tripdata_2018-01.csv')
```

```python
taxi_data = pd.read_csv("Documents/taxi _zone_lookup.csv")
taxi_data
```

```python
merged_data=pd.merge(yellow_cabs_data, taxi_data, on = None, left_on = 'PULocationID', right_on = 'LocationID')
df_Jan = pd.DataFrame(merged_data)
df_Jan
```

```python
#Cleaning: Removing all the values containing unknown

mergedata_jan = df_Jan[df_Jan.Borough != 'Unknown']
mergedata_jan
```

```python
#What is the most common way of payments

payment = mergedata_jan['payment_type'].value_counts().to_dict()
payment

# the most common way of payments is through Credit card(5996737)
```

```python
#Discover the way payments are executed in each borough
#initially, set that counter to 1.
mergedata_jan['COUNTER'] =1       
groupdata_jan = mergedata_jan.groupby(['payment_type','Borough'])['COUNTER'].sum() #sum function
print(groupdata_jan)
```

```python
#visualize the number of payments for any possible means.

ax = mergedata_jan[['payment_type','Borough']].plot(kind='bar', title ="Comparisions", figsize=(7, 5), legend=True, fontsize=5)
ax.set_xlabel("Payment Type", fontsize=5)
ax.set_ylabel("Borough", fontsize=5)
plt.show()
```

```python
#To check the relationship b/w Borough & payment type

paytyp_jan=mergedata_jan['payment_type']
borough_jan=mergedata_jan['Borough']

spearmanr_coefficient,p_value=spearmanr(paytyp_jan,borough_jan)
print ('Spearman Rank Correlation Coefficient %0.3f' %(spearmanr_coefficient))
```

```python
#Chi-square test

table=pd.crosstab(paytyp_jan,borough_jan)

from scipy.stats import chi2_contingency

chi2, p , dof, expected=chi2_contingency(table.values)

print 'Chi-square Statistics %0.3f' % (chi2,p)
```
